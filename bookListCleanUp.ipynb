```python
#import dependencies/libraries
import pandas as pd
import csv
```

```python
#declare file
file = "Resources/books.csv"
#read into data frame
# this line gives parse error due to varying row lengths booksDF = pd.read_csv(file)

#work around by putting into list then reread to data frame
#open file, do not forget encoding!
f_work = open(file, 'rt', encoding = 'utf-8')
#read it in
reader = csv.reader(f_work)
#put into a list
csv_list = []
for item in reader:
    csv_list.append(item)
#close file
f_work.close()
#read into data frame
booksDF = pd.DataFrame(csv_list)
#booksDF
```

```python
#drop last column
booksDF = booksDF.drop(columns = [12])
#rename columns
booksDF = booksDF.rename(columns = {0:"bookID", 1:"title", 2:"authors", 3:"average_rating", 4:"isbn", 5:"isbn13", 6:"language_code", 7:"num_pages", 8:"ratings_count", 9:"text_reviews_count", 10:"publication_date", 11:"publisher"})
#drop first row since that was the column name
booksDF = booksDF.drop([0])
booksDF
```
